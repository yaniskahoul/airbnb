# En fonction du type de room, calculez le nombre moyen d'amenities et l'écart type 

```python
df_listings_amsterdam['room_type'].unique()
```

```python
df_listings_amsterdam['amenities'].info()
```

```python
df_amenities = df_listings_amsterdam[["room_type","amenities"]]

df_amenities["number_of_amenities"] =df_amenities.amenities.str.count(',')
df_amenities["number_of_amenities"] =df_amenities.number_of_amenities.astype(int)
df_amenities["number_of_amenities"] = df_amenities.number_of_amenities.apply(lambda x:x+1)
df_amenities.groupby(df_amenities.room_type).number_of_amenities.agg(["mean","std"])
```

```python
df_bath= df_listings_amsterdam["bathrooms_text"]
df_bath = df_bath.dropna()
df_bath = df_bath.str.replace("baths","bath").str.replace("-"," ").str.replace("Private","private bath").str.replace("Bath","bath").str.replace("Half","half bath").str.replace("Shared","shared bath").str.replace("private bath", "2").str.replace("shared bath","0.5").str.replace("half bath","0.5").str.replace("bath","1")

df_listings_amsterdam[["num","type"]] = df_bath.str.split(" ",1,expand=True)

# df_bath.num.unique()
df_listings_amsterdam["num"] = df_listings_amsterdam["num"].astype("float")
df_listings_amsterdam["type"] = df_listings_amsterdam["type"].astype("float")
df_listings_amsterdam["result"] = df_listings_amsterdam["type"] * df_listings_amsterdam["num"]


df_listings_amsterdam.num.groupby(df_listings_amsterdam.result).count()








```
